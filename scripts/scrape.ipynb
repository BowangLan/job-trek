```python
import requests
from bs4 import BeautifulSoup
from rich import print
```

```python
url = "https://github.com/ReaVNaiL/New-Grad-2024/blob/main/README.md"
res = requests.get(url)
soup = BeautifulSoup(res.text)
```

```python
f"Number of tables: {len(soup.select('table'))}"
```

```python
table = soup.select("table")[0]
tbody = table.select("tbody")[0]
```

```python
row = tbody.find("tr")
row.td.find('a')
```

```python
data = []
def link_parser(item):
    return {
        "name": item.text.strip(),
        "link": item.find("a")['href'].strip("\"\\")
    }

def job_col_parser(item):
    jobs = [
        { 
            "name": j.text.strip(),
            "link": j['href'].strip("\"\\")
        }
        for j in item.find_all("a")
    ]
    text = item.text
    closed = False
    if "Closed" in text:
        closed = True
    return {
        "jobs": jobs,
        "open": not closed
    }
    
custom_parsers = {
    0: link_parser,
    2: job_col_parser
}
rows = tbody.select("tr")
for row in tbody.select('tr'):
    row_data = {}
    cols = row.select("td")
    company_name = cols[0].text.strip()
    company_job_url = cols[0].find("a")['href'].strip("\"\\")
    jobs = [
        { 
            "title": j.text.strip(),
            "url": j['href'].strip("\"\\")
        }
        for j in cols[2].find_all("a")
    ]
    closed = "Closed" in cols[2].text
    locations = [l.strip(" -") for l in cols[1].text.split("  - ")]
    note = cols[3].text.strip()
    data.append({
        "company": {
            "name": company_name,
            "jobUrl": company_job_url
        },
        "locations": locations,
        "jobs": jobs,
        "closed": closed,
        "note": note
    })
print(data[3:11])
```

```python
import json
import os
with open("../data/jobs.json", "w") as f:
    json.dump(data, f)
```
