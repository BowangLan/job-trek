```python
import requests
from bs4 import BeautifulSoup
from rich import print
```

```python
url = "https://github.com/ReaVNaiL/New-Grad-2024/blob/main/README.md"
res = requests.get(url)
soup = BeautifulSoup(res.text)
```

```python
f"Number of tables: {len(soup.select('table'))}"
```

```python
table = soup.select("table")[0]
tbody = table.select("tbody")[0]
```

```python
row = tbody.find("tr")
row.td.find('a')
```

```python
data = []
def link_parser(item):
    return {
        "name": item.text.strip(),
        "link": item.find("a")['href'].strip("\"\\")
    }

def job_col_parser(item):
    jobs = [
        { 
            "name": j.text.strip(),
            "link": j['href'].strip("\"\\")
        }
        for j in item.find_all("a")
    ]
    text = item.text
    closed = False
    if "Closed" in text:
        closed = True
    return {
        "jobs": jobs,
        "open": not closed
    }
    
custom_parsers = {
    0: link_parser,
    2: job_col_parser
}
rows = tbody.select("tr")
for row in tbody.select('tr'):
    row_data = []
    for i,item in enumerate(row.select("td")):
        if i in custom_parsers:
            row_data.append(custom_parsers[i](item))
        else:
            row_data.append(item.text.strip())
    data.append(row_data)
print(data[3:11])
```
